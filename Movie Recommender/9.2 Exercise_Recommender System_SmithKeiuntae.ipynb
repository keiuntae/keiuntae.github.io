```python
# Load libraries
import pandas as pd
import numpy as np
```

## Load the Datasets that will be used in this exercise

```python
# load the ratings dataset and preview the first 10 rows
data = pd.read_csv('ratings.csv')
data.head(10)
```

```python
# load the movie titles dataset and preview the first 10 rows
movie_titles_genre = pd.read_csv("movies.csv")
movie_titles_genre.head(10)
```

```python
# Merge the two dataframes into one 
data = data.merge(movie_titles_genre,on='movieId', how='left')

# Preview the first 10 rows of the new dataframe
data.head(10)
```

```python
#create a new dataframe that displays average rating (mean) for each movie in the data
mean_rating = pd.DataFrame(data.groupby('title')['rating'].mean())
mean_rating.head(10)
```

## Total Number of Ratings

```python
#create a dataframe that shows the total ratings cast for each movie
mean_rating['Total Ratings'] = pd.DataFrame(data.groupby('title')['rating'].count())
mean_rating.head(10)
```

## Correlation Calculation 

```python
#Calculating The Correlation by making a pivot table (rows=userID, columsns=Movie Title)
movie_user = data.pivot_table(index='userId',columns='title', values='rating')
```

```python
# Preview the first ten rows pivot dataframe
movie_user.head(10)
```

## Movie Selection

```python
#create a correlation value using the 'corrwith' function in conjunction with a movie choice
correlations = movie_user.corrwith(movie_user['Godfather, The (1972)'])

#display the first five rows of computed pairwise correlation between rows and columns
correlations.head()
```

```python
# create a new variable named 'recommend' and drop all the empty values 
recommend = pd.DataFrame(correlations,columns=['Correlation'])
recommend.dropna(inplace=True)

#Merge the ratings to the correlation table 
recommend = recommend.join(mean_rating['Total Ratings'])

#Preview the first 5 rows
recommend.head()
```

```python
#filter all movies that are correlated to 'Godfather, The (1972)' using the sort_values function
recc = recommend[recommend['Total Ratings']>100].sort_values('Correlation',ascending=False).reset_index()
```

```python
# merge the original movie dataset to show all fields 
recc = recc.merge(movie_titles_genre,on='title', how='left')
```

```python
#display the recommended list dataframe to include the movie selection and ten recommended movies
recc.head(11)
```
